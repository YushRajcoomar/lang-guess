```python
from datasets import load_dataset
import soundfile as sf
import os
import random

# Create main output directory
output_dir = "/Users/yushrajcoomar/Desktop/projects/languages_game/language-guessr-3d/client/src/assets/audio"
os.makedirs(output_dir, exist_ok=True)

# List of languages with ISO codes and country codes
languages = [
    # {"name": "French", "iso": "fr_fr", "country": "FRA"},
    # {"name": "Spanish", "iso": "es_419", "country": "ESP"},
    # {"name": "Mandarin", "iso": "cmn_hans_cn", "country": "CHN"},
    # {"name": "Arabic", "iso": "ar_eg", "country": "EGY"},
    # {"name": "Hindi", "iso": "hi_in", "country": "IND"},
    # {"name": "German", "iso": "de_de", "country": "DEU"},
    # {"name": "Japanese", "iso": "ja_jp", "country": "JPN"},
    # {"name": "Portuguese", "iso": "pt_br", "country": "BRA"},
    # {"name": "Russian", "iso": "ru_ru", "country": "RUS"},
    # {"name": "Swahili", "iso": "sw_ke", "country": "KEN"}
]

# Extract samples for each language
for lang in languages:
    print(f"Processing {lang['name']}...")
    
    # Create language-specific subdirectory
    lang_dir = os.path.join(output_dir, lang['name'].lower())
    os.makedirs(lang_dir, exist_ok=True)
    
    # Load just this language's data
    dataset = load_dataset("google/fleurs", lang['iso'], split="validation")
    
    # Select random samples
    samples = random.sample(range(len(dataset)), 5)
    
    for i, idx in enumerate(samples):
        # Get audio array and sampling rate
        audio = dataset[idx]["audio"]["array"]
        sampling_rate = dataset[idx]["audio"]["sampling_rate"]
        
        # Save as wav file in the language subdirectory
        output_file = os.path.join(lang_dir, f"{i+1}.wav")
        sf.write(output_file, audio, sampling_rate)
        
        print(f"  Saved {output_file}")

print("Done extracting audio samples!")
```

```python
from datasets import load_dataset, get_dataset_config_names
import soundfile as sf
import os
import random
import numpy as np
import itertools

# Create main output directory
output_dir = "/Users/yushrajcoomar/Desktop/projects/languages_game/language-guessr-3d/client/src/assets/audio"
os.makedirs(output_dir, exist_ok=True)

# Define the RMS normalization function
def normalize_audio(audio, target_rms=0.1):
    """Normalize audio using RMS normalization to ensure consistent loudness"""
    # Calculate current RMS
    current_rms = np.sqrt(np.mean(audio**2))
    
    # Calculate required gain
    gain = target_rms / (current_rms + 1e-6)  # Adding small value to prevent division by zero
    
    # Apply gain (with clipping protection)
    normalized_audio = audio * gain
    
    # Clip to prevent distortion
    if np.max(np.abs(normalized_audio)) > 1.0:
        normalized_audio = normalized_audio / np.max(np.abs(normalized_audio))
    
    return normalized_audio

# Extract samples for each language
for lang in updated_data:
    # Skip languages that aren't available
    print(f"Processing {lang['country_name']}...")
    
    # Create language-specific subdirectory
    country_name = lang['country_name'].lower()
    if ' ' in country_name:
        country_name = country_name.replace(' ', '_')
    lang_dir = os.path.join(output_dir, country_name)
    os.makedirs(lang_dir, exist_ok=True)
    
    try:
        # Load this language's data with streaming
        dataset = load_dataset("google/fleurs", lang['iso'], split="validation", streaming=True)
        
        # For streaming datasets, we need a different approach to get random samples
        # First, we'll convert the iterable to a list of the first 20 examples (not loading all to memory)
        sample_pool = list(itertools.islice(dataset, 20))
        
        # Now we can select random samples from this smaller pool
        num_samples = min(5, len(sample_pool))
        selected_samples = random.sample(sample_pool, num_samples)
        
        for i, sample in enumerate(selected_samples):
            # Get audio array and sampling rate
            audio = sample["audio"]["array"]
            sampling_rate = sample["audio"]["sampling_rate"]
            
            # Apply RMS normalization
            normalized_audio = normalize_audio(audio)
            
            # Save as wav file in the language subdirectory
            output_file = os.path.join(lang_dir, f"{i+1}.wav")
            sf.write(output_file, normalized_audio, sampling_rate)
            
            print(f"  Saved {output_file}")
            
    except Exception as e:
        print(f"  Error processing {lang['country_name']}: {str(e)}")

print("Done extracting audio samples!")
```
